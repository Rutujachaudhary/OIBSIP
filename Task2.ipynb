```python
#import the libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Load the dataset
df = pd.read_csv('spam.csv', encoding='ISO-8859-1')
```

```python
df
```

```python
df.info()
```

```python
df.head(5)#to get the top 5 records
```

```python
df.shape
```

```python
df.drop(columns=['Unnamed: 2','Unnamed: 3','Unnamed: 4'],inplace=True)#remove unwanted column
```

```python
df
```

```python
df=df.rename(columns={'v1':'Category','v2':'Message'})
```

```python
df.head(5)
```

```python
data=df.where((pd.notnull(df)),'')
```

```python
data.head(5)
```

```python
data.info()
```

```python
data.loc[data['Category']=='spam','Category',]= 0
data.loc[data['Category']=='ham','Category',]= 1
```

```python
X=data['Message']
Y=data['Category']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
print(Y.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train)
```

```python
print(X_train_features)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train_features,Y_train)
```

```python
prediction_on_training_data =  model.predict(X_train_features)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
```

```python
print('Acc on training data : ', accuracy_on_training_data)
```

```python
prediction_on_test_data = model.predict(X_test_features)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
```

```python
print('acc on test data : ',accuracy_on_test_data)
```

```python
input_your_mail = ["Congrats! Nokia 3650 video camera phone is your Call 09066382422 Calls cost 150ppm Ave call 3mins vary from mobiles 16+ Close 300603 post BCM4284 Ldn WC1N3XX"]

input_data_features = feature_extraction.transform(input_your_mail)

prediction = model.predict(input_data_features)

print(prediction)

if(prediction[0]==1):
   print('Ham mail')

else:
  print('Spam mail')
```
