Aanalyzing Data

```python
# Group data by Region and calculate mean of Estimated Unemployment Rate
grouped_data = df.groupby('Region')[' Estimated Unemployment Rate (%)'].mean()

plt.figure(figsize=(12, 6))
colors = plt.cm.tab20c.colors
bars = plt.bar(grouped_data.index, grouped_data.values, color=colors)
plt.title('Unemployment Rate (State-Wise) by Bar Graph')
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python

# Group data by Region and calculate mean of Estimated Unemployment Rate
grouped_data = df.groupby('Region')[' Estimated Unemployment Rate (%)'].mean()

# Sort data in ascending order of 'Estimated Unemployment Rate (%)'
sorted_data = grouped_data.sort_values()
plt.figure(figsize=(12, 6))
plt.bar(sorted_data.index, sorted_data.values, color=plt.cm.Paired.colors)
plt.title('Unemployment Rate (State-Wise) by Bar Graph')
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python

# Group data by Region.1 and calculate mean of Estimated Unemployment Rate
grouped_data = df.groupby('Region.1')[' Estimated Unemployment Rate (%)'].mean()

# Sort data in descending order of mean Estimated Unemployment Rate
sorted_data = grouped_data.sort_values(ascending=False)
plt.figure(figsize=(10, 6))
bars = plt.bar(sorted_data.index, sorted_data.values, color='lightgreen')
plt.title('Unemployment Rate (Region-Wise) by Bar Graph')
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend(bars, sorted_data.index, loc='best')

plt.show()
```

```python
import plotly.express as px
fg=px.scatter(df,x='Region',y=' Estimated Unemployment Rate (%)',color='Region',
          title='Unemployment Rate (State-Wise) by Scatter Graph',template='plotly')
fg.update_layout(xaxis={'categoryorder':'total descending'})
fg.show()
```
